```python
import pandas as pd
from sklearn import preprocessing 
import numpy as np
```

```python
csv_file = '6TaipeiObserve.csv'
train_set = pd.read_csv(csv_file)
train_set.head()
```

```python
def normalization(df, key):
    """Normalize a dataframe with specified columns

    Keyword arguments:
    df -- the input dataframe (pandas.DataFrame)
    cols -- the specified columns to be normalized (list)

    """
    x = df[[key]].values.astype(float)
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    normalized_data = pd.DataFrame(x_scaled)
    normalized_data.columns = [key]
    return normalized_data
```

```python
normalize_train_set = train_set[['觀測時間']].copy()
```

```python
keys = ['測站氣壓', '海平面氣壓', '氣溫', '露點溫度', '相對溼度', '風速']
for key in keys:
    temp = normalization(train_set, key)
    normalize_train_set = pd.concat([normalize_train_set, temp], axis=1, sort=False)
```

```python
#train_set.降水量 = train_set.降水量.convert_objects(convert_numeric=True)
normalize_train_set['降水量'] = np.where(train_set.降水量 > 0.5, 1, 0)
```
